```python
def autopct_format(value):
    def my_format(pot):
        total = sum(values)
        val = int(round(pot * total/100.0))
        return '$ (vid)'.format(v=val)
```

```python
#Total de vendas por segmento e por ano
df_st['Data do pedido'] = pd.to_datetime(df_st['Data do pedido'], dayfirst = True)
```

```python
#Adicionando a coluna ano
df_st['Ano']= df_st['Data do pedido'].dt.year
```

```python
df_st.head(3)
```

```python
df_st_t6 = df_st.groupby(['Ano','Categoria'])['Vendas'].sum()
```

```python
#descontos venda>1000=15% desc; venda<1000=10% desc
df_st['Desconto'] = np.where(df_st['Vendas']>1000, 0.15, 0.10)
```

```python
df_st.head()
```

```python
df_st['Desconto'].value_counts()
```

```python
#média valor de vendas antes e depois dos descontos 
```

```python
df_st['venda_desc'] = df_st['Vendas'] - (df_st['Vendas'] * df_st['Desconto'])
```

```python
df_st.head()
```

```python
df_st_pre_desc = df_st.loc[df_st['Desconto'] == 0.15, 'Vendas']
```

```python
df_st_after_desc = df_st.loc[df_st['Desconto'] == 0.15, 'venda_desc']
```

```python
media_venda_pre_desc = df_st_pre_desc.mean()
```

```python
media_venda_pos_desc = df_st_after_desc.mean()
```

```python
print('Média de vendas sem desconto', round(media_venda_pre_desc, 2))
```

```python
print('Média de vendas com desconto ', round(media_venda_pos_desc, 2))
```
